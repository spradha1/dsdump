```python
import pandas as pd
import numpy as np
```

```python
# data

df = pd.read_csv('data.csv', index_col=False)
# remove unnamed column
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
print(f'Shape: {df.shape}')
df.head()
```

```python
print(df.columns)
df.dtypes.unique()
```

```python
df.describe()
```

```python
# just string types
df.describe(include=object)
```

```python
# null values
df.isna().sum(axis=0).sort_values()
```

```python
# impute numerical columns with mean and categorical ones with mode

cols = df.columns
na_types = set()
for c in cols:
  # if at least one NA is present in the column
  if df[c].isna().any():
    if df[c].dtype == object:
      df[c].fillna(df[c].mode(), inplace=True)
    else:
      df[c].fillna(df[c].mean(), inplace=True)
    if df[c].isna().any():
      na_types.add(df[c].dtype)

print(na_types)
# for the rest of the columns with data type 'dtype('O')
df.fillna(0, inplace=True)
df.isna().sum().any()
```

```python
# Wage column should be refined
df.Wage.unique()
```

```python
# create integer mappings for string wages in the dataframe
num_wages = dict()
for w in np.array(df.Wage.unique()):
  nw = []
  for ch in w:
    # check for characters that are integers
    try:
      nw.append(str(int(ch)))
    except:
      continue
  num_wages[w] = int(''.join(nw))
print(num_wages)
```

```python
# refine data with the wage mapping and incorporate into dataframe

df.Wage = df.Wage.apply(lambda x: num_wages[x])
df.Wage.unique()
```

```python
# write out clean data to new file

df.to_csv('data_clean.csv', index=False)
```
