```python
import pandas as pd
```

```python
df = pd.read_csv('../datasets/weatherAUS.csv')
df.head()
```

```python
print(df.shape)
print(df.dtypes)
df.describe()
```

```python
df.isna().sum(axis=0)
```

```python
df1 = df.copy()
df1.columns
```

```python
# data imputation

# median for numeric columns
df1.MinTemp.fillna(df1.MinTemp.median(), inplace=True)
df1.MaxTemp.fillna(df1.MaxTemp.median(), inplace=True)
df1.Rainfall.fillna(df1.Rainfall.median(), inplace=True)
df1.Evaporation.fillna(df1.Evaporation.median(), inplace=True)
df1.Sunshine.fillna(df1.Sunshine.median(), inplace=True)
df1.WindGustSpeed.fillna(df1.WindGustSpeed.median(), inplace=True)
df1.WindSpeed9am.fillna(df1.WindSpeed9am.median(), inplace=True)
df1.WindSpeed3pm.fillna(df1.WindSpeed3pm.median(), inplace=True)
df1.Humidity9am.fillna(df1.Humidity9am.median(), inplace=True)
df1.Humidity3pm.fillna(df1.Humidity3pm.median(), inplace=True)
df1.Pressure9am.fillna(df1.Pressure9am.median(), inplace=True)
df1.Pressure3pm.fillna(df1.Pressure3pm.median(), inplace=True)
df1.Cloud9am.fillna(df1.Cloud9am.median(), inplace=True)
df1.Cloud3pm.fillna(df1.Cloud3pm.median(), inplace=True)
df1.Temp9am.fillna(df1.Temp9am.median(), inplace=True)
df1.Temp3pm.fillna(df1.Temp3pm.median(), inplace=True)

# mode for string columns
df1.WindGustDir.fillna(df1.WindGustDir.mode().values[0], inplace=True)
df1.WindDir9am.fillna(df1.WindDir9am.mode().values[0], inplace=True)
df1.WindDir3pm.fillna(df1.WindDir3pm.mode().values[0], inplace=True)
df1.RainToday.fillna(df1.RainToday.mode().values[0], inplace=True)
```

```python
df1.isna().sum()
```

```python
# save clean csv file
df1.to_csv('../datasets/weatherAUS_clean.csv', index=False) # index False to not write index columns
```
